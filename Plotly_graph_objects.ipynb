```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
import numpy as np
import  seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=px.data.gapminder()
df.head(20)
```

```python
fig=go.Figure()
fig.show()
```

```python
fig=go.Figure(go.Bar(x=df['country'],y=df['pop']))
fig.show()
```

Let's add a "trace" to the object figure.

```python
df1=df.query('year==2007')
fig=go.Figure(go.Bar(x=df1['country'],y=df1['pop']))
fig.show()
fig.write_html("bar.html")
```

Let's add the layout. What is the plot supposed to look like?

```python
fig=go.Figure(data=go.Bar(x=df1['country'],y=df1['pop']),
              layout=go.Layout(title=go.layout.Title(text="Population of Countries"))
              )
fig.show()
```

```python
fig=go.Figure(data=go.Bar(
                          x=df1['country'],y=df1['pop']
                          ),
              layout=go.Layout(
                      title_text="Population of Countries",
                      xaxis_title_text="Country Name",
                      yaxis_title_text="Population"
                      )
              )
fig.show()
```

Let's add another plot - "trace" to the figure.

```python
df2=df.query('year==1952')
fig=go.Figure(data=[go.Bar(
                          x=df1['country'],y=df1['pop'],
                          ),
                    go.Bar(
                          x=df2['country'],y=df2['pop']
                          )
                    ],
              layout=go.Layout(
                      title_text="Population of Countries",
                      xaxis_title_text="Country Name",
                      yaxis_title_text="Population",
                      legend_title_text="Year"
                      )
              )
fig.show()
```

```python
df2=df.query('year==1952')
fig=go.Figure(data=[ go.Bar(
                          x=df1['country'],y=df1['pop'],
                          name="2007"
                          ),
                go.Bar(
                          x=df2['country'],y=df2['pop'],
                          name="1952"
                          )
                    ],
              layout=go.Layout(
                      title_text="Population of Countries",
                      xaxis_title_text="Country Name",
                      yaxis_title_text="Population",
                      legend_title_text="Year"
                      )
              )
fig.show()
```

```python
df3=df.query('year==1982')
fig.add_trace(go.Bar(
                          x=df3['country'],y=df3['pop'],
                          name="1982"
                          ))
fig.update_traces()
fig.show()
```

Since we are looking at different countries, let's make some maps.

```python
fig_map=px.scatter_geo(df1,locations='iso_alpha')
fig_map.show()
```

```python
fig_map=px.scatter_geo(
    df1,
    locations='iso_alpha',
    hover_name='country',
    color='continent',
    size='pop'
    )
fig_map.show()
```

```python
fig_map=px.scatter_geo(
    df1,
    locations='iso_alpha',
    hover_name='country',
    color='continent',
    size='pop',
    projection='natural earth'
    )
fig_map.show()
```

```python
fig_map=px.scatter_geo(
    df,
    locations='iso_alpha',
    hover_name='country',
    color='continent',
    size='gdpPercap',
    projection='natural earth',
    animation_frame='year',
    title='Population of Countries'
    )
fig_map.show()

fig_map.write_html("map.html")
```

```python
fig_map=px.choropleth(df1,
                      locations='iso_alpha',
                      color='pop',
                      hover_name='country')
fig_map.show()
```
