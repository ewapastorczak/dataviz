```python
#Empty application
app = Dash(__name__)

#HTML layout
app.layout = html.Div([
    "Pick a Country",
    dcc.Dropdown(
        options=["Poland", "Spain", "France", "Indonesia", "Nigeria"],
        value="Poland",
        id="country-dropdown"
    ),
    #header 3rd level
    html.H3(id="country-output")
])


if __name__ == "__main__":
    app.run(port=8060)
```
