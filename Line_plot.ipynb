```python
#essential data analysis
import numpy as np
#friendly data exploration
import pandas as pd
#essential plot production
import matplotlib.pyplot as plt
#matplotlib but pretty
import seaborn as sns
#matplotlib but modern
import plotly
```

```python
df=sns.load_dataset('fmri')
print(df.head())
df.describe(include='all')
```

```python
sns.relplot(data=df,x='timepoint',y='signal')
```

```python
sns.relplot(data=df,x='timepoint',y='signal',kind='line')
```

```python
sns.relplot(data=df,x='timepoint',y='signal',kind='line',estimator=None)
```

```python
sns.relplot(data=df,x='timepoint',y='signal',kind='line',style='region',markers=True,hue='region')
```

```python
sns.relplot(data=df,x='timepoint',y='signal',kind='line',style='region',markers=True,hue='region',col='subject',col_wrap=2)
```

```python
sns.jointplot(data=df,x='timepoint',y='signal',kind='scatter')
```

```python
sns.jointplot(data=df,x='timepoint',y='signal',kind='hex')
```

```python
sns.jointplot(data=df,x='timepoint',y='signal',kind='kde',n_levels=20,cmap='coolwarm')
```

```python
sns.jointplot(data=df,x='timepoint',y='signal',kind='reg')
```
