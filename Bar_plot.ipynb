```python
df=sns.load_dataset('tips')
df.head()
```

```python
sns.catplot(data=df,x='day',y='total_bill',kind='bar')
```

```python
sns.catplot(data=df,x='day',y='total_bill',kind='bar',errorbar=None,hue='day',palette='Blues')
```

```python
sns.catplot(data=df,x='total_bill',y='day',kind='bar',errorbar=None,hue='day',palette='Blues')
```

```python
sns.catplot(data=df,x='total_bill',y='day',kind='bar',errorbar=None,hue='sex',palette='Blues')
```

Seaborn stacked barplot is quite tedious to create. If you insist, you can use pandas+a seaborn style.

```python
stack_df=df.groupby(['day', 'sex']).total_bill.mean().unstack()
print(stack_df)
stack_df.plot(kind='bar', stacked=True, color=['tab:blue', 'tab:cyan'], width=0.8)
#(df
#    .groupby(['day', 'sex'])
#    .['total_bill'
#    .sum()
#    .unstack()
#).plot(kind='bar', stacked=True, color=['tab:blue', 'tab:cyan'], width=0.8)
```
