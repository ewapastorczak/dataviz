```python
#essential data analysis
import numpy as np
#friendly data exploration
import pandas as pd
#essential plot production
import matplotlib.pyplot as plt
#matplotlib but pretty
import seaborn as sns
#matplotlib but modern
import plotly
```

```python
df=sns.load_dataset('flights')
df.head()
```

```python
sns.lineplot(x='year',y='passengers',data=df,estimator=sum,errorbar=None)
```

```python
sns.barplot(x='month',y='passengers',data=df,estimator=np.mean,errorbar=None,palette='rocket',hue='month')
```

**Heat map charts**

a grid or matrix where individual cells are colored according to the data values they represent

**Why use a heat map?**


* Identify a patter: quickly spot correlations, patterns, or anomalies in your data.
* Compare data across categories: e.g. sales figures across different regions and time periods.
* Highlight highs and lows
* Visualize a large dataset: When dealing with large amounts of data, a heat map chart can condense the information into a more digestible format, allowing for faster insights.

image.png


We now want to make a heatmap, with axes representing years and months and colors representing the number of passengers.

We need to do a minor data processing.

```python
df1=df.pivot(index='month',columns='year',values='passengers')
df1.head()
```

```python
sns.heatmap(df1)
#
```

```python
sns.set(font_scale=0.75)
sns.heatmap(df1,annot=True,fmt='d',cmap='coolwarm')
```
