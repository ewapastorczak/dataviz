```python
#essential data analysis
import numpy as np
#friendly data exploration
import pandas as pd
#essential plot production
import matplotlib.pyplot as plt
#matplotlib but pretty
import seaborn as sns
#matplotlib but modern
import plotly
```

```python
df=sns.load_dataset('iris')
print(df.head())
df.describe(include='all')
```

image.png

```python
sns.pairplot(df)
```

Good for:

- initial look at the data


Usually bad for:

- explanatory storytelling

```python
sns.pairplot(df,hue='species')
```

Making your own customary pairplot - The Pair Grid class

```python
g=sns.PairGrid(df)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot)
#
```

```python
g=sns.PairGrid(df,hue='species',palette='coolwarm')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot)
```

Finally!

**Linear regression**

https://seaborn.pydata.org/tutorial/regression.html

```python
sns.regplot(data=df,x='petal_width',y='petal_length')
```

```python
sns.lmplot(data=df,x='petal_width',y='petal_length')
```

```python
sns.lmplot(data=df,x='petal_width',y='petal_length',hue='species')
```


Non-linear function fit


```python
sns.lmplot(data=df,x='petal_width',y='petal_length',hue='species',order=2,ci=None)
```

Robust regression when you have outliers

```python
sns.lmplot(data=df,x='petal_width',y='petal_length',hue='species',ci=None,robust=True)
```

Logistic regresssion for binary variables

image.png

```python
sns.lmplot(data=df,x='sepal_width',y='petal_length',ci=None,logistic=True)
```
